```
import pandas as pd # importing pandas in order to read the csv file
df = pd.read_csv('https://raw.githubusercontent.com/frankData612/data_612/master/State_Drug_Utilization_Data_2010/State_Drug_Utilization_Data_2010.csv')
df.head(10) # using .head() to show only the first rows of the dataframe. df.head(10) is used to see the first 10 rows.
```

```
# df.groupby will group by the selected feature. In this instance the quarter is the group and the mean function is being applied to the number of prescriptions in each quarter.
pres_mean = df.groupby('Quarter')['Number of Prescriptions'].mean()
pres_mean
```
