## About the assignment

In the industry, analytics professionals often explore data's measure variables grouped by one or multiple dimension variables. For example, the metrics of interest could be the average, max, min, or percentage (mix) of sales by month, location, etc.

The followings are the sample code snippet using the custom user functions together with `groupby()` and `apply()` in another way:


```
import pandas as pd

# the mpg data is used for demonstration
df = pd.read_csv("https://raw.githubusercontent.com/franklin-univ-data-science/data/master/mpg.csv")
df.head()
```

```
def summarize_data(g):
  return(
    pd.Series({
      "avg_MPG": g.mpg.mean(),
      "min_MPG": g.mpg.min(),
      "max_MPG": g.mpg.max(),
      "sample_count": len(g),
      "mix": g.cylinders.sum()/df.cylinders.sum(),
    })
  )

df.groupby("cylinders").apply(summarize_data)
```

```
import pandas as pd # pandas used to read csv as well as part of the function

prescriptions = pd.read_csv('https://raw.githubusercontent.com/frankData612/data_612/master/State_Drug_Utilization_Data_2010/State_Drug_Utilization_Data_2010.csv')

def summarize_prescriptions(g): # function defined for several series and statistics
  return(
    pd.Series({
      "avg_prescriptions": g['Number of Prescriptions'].mean(), # column name and values are defined in this dictionary
      "min_prescriptions": g['Number of Prescriptions'].min(),
      "max_prescriptions": g['Number of Prescriptions'].max(),
      "sample_count": len(g),
      "mix": g['Number of Prescriptions'].sum()/prescriptions['Number of Prescriptions'].sum(),
    })
  )

prescriptions.groupby("State").apply(summarize_prescriptions) # grouping by State and running the function to get a summary for each State
```
