```
import seaborn as sns   # Importing seaborn and matplotlib.pyplot for plotting 
import pandas as pd     # tools and pandas to manipulate the dataframe
import matplotlib.pyplot as plt

sns.get_dataset_names() # used to find an interesting dataset to plot
```

```
peng = sns.load_dataset('penguins') # With seaborn, the penguins dataset was imported

peng['species'].unique() # Determined the different species of penguin
```

```
adelie = peng[peng['species'] == "Adelie"] # Decided to include only the Adelie species to plot
adelie
```

```
hist, ax = plt.subplots() # histograms work best to analyze a single variable

ax = sns.distplot(adelie['bill_length_mm'], kde= False) # seaborn distplot creates a histogram only when kde = False

ax.set_title('Adelie Penguin Bill Length Distribution') # title set

ax.set_xlabel('Bill Length(mm)') # axes labels set

ax.set_ylabel('Count')

plt.show() # Final result shows a somewhat normal distribution of bill lengths
```

```
scatter, ax = plt.subplots() # Using the same method a scatterplot is made with a regression line

ax = sns.regplot(x = 'bill_length_mm', y = 'body_mass_g', data = adelie)

ax.set_title('Bill Length vs Body Mass in Adelie Penguins')

ax.set_xlabel('Bill Length(mm)')

ax.set_ylabel('Body Mass(g)')

plt.show() # scatter plots are best used to compare two continuous variables
```

```
hexbin = sns.jointplot(x="bill_length_mm", y="body_mass_g", data = adelie, kind="hex") # joint plots group close data points together in a bin

hexbin.set_axis_labels(xlabel='bill_length_mm', ylabel='body_mass_g') # the hex was chosen as the shape of the bins

hexbin.fig.suptitle('Hexbin Joint Plot of Bill Length and Body Mass of Adelie Penguins', # a histogram is also included for each variable on the axes

                 fontsize=10, y = 1) # joint plots are best when there are many data points clustered together making distinction between them difficult
```
