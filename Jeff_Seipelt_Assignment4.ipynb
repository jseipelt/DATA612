```
import seaborn as sns   # Importing seaborn to access the penguin dataset
import pandas as pd     # tools and pandas to manipulate the dataframe
sns.get_dataset_names()
```

```
peng = sns.load_dataset('penguins') # Taking the penguin dataset and will separate it as an example for merge
peng
```

```
peng_1 = peng[peng['species'] == 'Adelie'] # The first dataframe will use the Adelie species

peng_2 = peng[peng['species'] == 'Gentoo'] # The second dataframe uses teh Gentoo species
peng_2
```

```
peng_merge = peng_1.merge(peng_2, # Used merge for each column with an outer join since there would be no duplicate values
                          left_on=['species','island','bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g','sex'],
                          right_on=['species','island','bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g','sex'],
                          how='outer')

peng_merge
```

```
peng_merge.count() # to find the number of missing values, begin by counting the total number of non-missing values for each column
```

```
num_rows = peng_merge.shape[0] # Next count total number of rows
```

```
num_rows - peng_merge.count() # subtracting the number of non-missing values from the total shows the number of missing values for each column
```

```
peng_merge.fillna(method='ffill') # Cleaning the data by filling in the values that come before the missing values removes them all
```
