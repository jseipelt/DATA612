```
import pandas as pd # importing pandas as pd to read the csv

drug_data = pd.read_csv('https://raw.githubusercontent.com/frankData612/data_612/master/State_Drug_Utilization_Data_2010/State_Drug_Utilization_Data_2010.csv')
drug_data.tail()
```

```
import re #importing regular expression module

prod_name = drug_data["Product Name"] # defining the 'Product Name' column
drug_data['Product Name_cleaned'] =  [re.sub(r'[(][a-zA-Z]?','', str(x)) for x in prod_name] # Using the sub function, '(' is replaced by '' 
                                                                                             # as well as if there are any letters following the '('
drug_data.tail()                                                                             # ALLEGRA (F is converted to only ALLEGRA and BACTROBAN( to BACTROBAN
```

```
import statistics

def stats(x): # Creating a function to call mean, sum, mode, median and range
  
  x = drug_data['Number of Prescriptions']

  mean = sum(x) / len(x) # defining mean as the sum of the column divided by the total number of values in the column
  sum_stat = sum(x)
  mode = statistics.mode(x)
  median = x[len(x)//2] # taking the middle number using the len function and dividing by 2
  range = max(x) - min(x)
  return print('Mean:', mean, ',', 'Sum:', sum_stat, ',', 'Mode:', mode, ',', 'Median:', median, ',', 'Range:', range)

x.apply(stats) # Using .apply causes the stats to print for every index instead of only once
```
